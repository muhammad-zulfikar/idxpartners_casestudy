Import the necessary libraries.

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

Load the customer data.

```python
df = pd.read_csv('/workspaces/IDXPartnersCaseStudy/data/bank.csv')
```

 Select the features that you want to use for clustering.

```python
features = ['balance', 'age', 'campaign']
```

Create the K-Means model, fit the K-Means model to the data, and predict the cluster labels for each customer.

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(df[features])
labels = kmeans.predict(df[features])
```

Visualize the clusters

```python
plt.scatter(df["balance"], df["age"], c=labels)
plt.show()
```
